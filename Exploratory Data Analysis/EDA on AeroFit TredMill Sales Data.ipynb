# DASHBOARD

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact



# Function to determine if a column is categorical or numerical
def is_categorical(series):
    return series.dtype == 'object' or len(series.unique()) < 10

# Function for enhanced dashboard
def plot_dashboard(feature):
    plt.figure(figsize=(10, 6))

    if is_categorical(df[feature]):
        sns.countplot(x=feature, data=df, palette='coolwarm')
        plt.title(f"Count Plot of {feature}")
        plt.xlabel(feature)
        plt.ylabel('Count')
    else:
        sns.histplot(df[feature], kde=True, color='blue', bins=10)
        plt.title(f"Distribution of {feature}")
        plt.xlabel(feature)
        plt.ylabel('Frequency')

    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Interactive Widget with dynamic column selection
interact(plot_dashboard, feature=df.columns)

```

#Outlier Detection

```python
# IQR method for detecting outliers
for column in numeric_df:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1 # FORMULA
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"Outliers in {column}:")
    print(df[(df[column] < lower_bound) | (df[column] > upper_bound)])
```

```python
from scipy.stats import zscore

# Z score for outliears
z_scores = numeric_df.apply(zscore)

# Find outliers
outliers = (z_scores > 3) | (z_scores < -3)
print("Outliers Detected:")
print(df[outliers.any(axis=1)])
```

```python
# Box plot for Outliers
for column in numeric_df:
    sns.boxplot(x=df[column])
    plt.title(f"Box Plot for {column}")
    plt.show()
```

**Recommendations:**
- **Premium marketing focus**: Highlight advanced features of KP781 (e.g., smart connectivity, personalized fitness tracking) in campaigns targeting high-income customers.
- **Exclusive offers**: Introduce loyalty programs or luxury upgrades (e.g., extended warranties, premium add-ons) for high-income buyers.
- **High-end distribution**: Make KP781 available in premium retail locations or through exclusive online channels.

--------------------------------------------------------------------------------


### **2. Fitness Level 5 Customers**
**Observations:**
- **17.22% of total customers** have Fitness Level 5, indicating the highest fitness level.
- Among these, **16.11% purchase Product KP781**.
- There is a moderate uptake of KP781 among highly fit customers.

**Recommendations:**
- **Fitness-focused promotions**: Market KP781 as the ultimate treadmill for fitness enthusiasts, emphasizing features like endurance tracking and high-performance design.
- **Engage fitness influencers**: Collaborate with highly fit influencers to promote KP781 on social platforms.
- **Bundle offers for enthusiasts**: Offer packages that include KP781 along with fitness apps, wearables, or nutrition plans to appeal to Fitness Level 5 customers.

------------------------------------------------------------------------------
### **3. Partnered Customers Using Treadmills**
**Observations:**
- **59.44% of partnered customers** use treadmills, representing a significant portion of the customer base.
- This indicates strong engagement of partnered individuals with treadmill products.

**Recommendations:**
- **Family-friendly campaigns**: Promote treadmills as part of family fitness goals, emphasizing health and bonding for partnered customers.
- **Group discounts**: Provide special deals for couples or families purchasing multiple products.
- **Wellness programs**: Partner with fitness centers to offer partnered customers discounted gym memberships when they purchase treadmills.
--------------------------------------------------------------------------------

### **4. Gender Analysis**
**Observations:**
- KP781 is more popular among **female customers**, while KP281 and KP481 have relatively balanced gender preferences.
- Male customers tend to purchase treadmills more frequently overall.

**Recommendations:**
- **Tailored promotions**: Use gender-based marketing strategies to target male customers with strength-focused features and female customers with health and lifestyle benefits.
- **Expand female customer base**: Offer treadmills in colors or designs that appeal more to women.

--------------------------------------------------------------------------------


### **5. Age Analysis**
**Observations:**
- Customers in their **20s and 30s** dominate treadmill purchases, while engagement from customers aged 40+ is lower.

**Recommendations:**
- **Target younger demographics**: Emphasize modern designs, app integrations, and fitness challenges that appeal to millennials.
- **Attract older customers**: Introduce treadmills with features like joint protection and ease of use to attract older demographics.

--------------------------------------------------------------------------------

### **6. Income Analysis**
**Observations:**
- **Low-income customers** prefer entry-level models like KP281.
- **High-income customers** show a preference for premium products like KP781.

**Recommendations:**
- **Diversify product offerings**: Maintain a range of treadmills catering to different income brackets.
- **Flexible financing**: Provide installment payment options for low-income customers to boost their ability to purchase.

--------------------------------------------------------------------------------


### **7. Correlation Analysis**
**Observations:**
- Strong correlations exist between product preference and fitness level, income, and gender.
- Weak correlations were found between marital status and product choice.

**Recommendations:**
- **Leverage key features**: Focus on high-correlation factors (income and fitness level) in product development and marketing strategies.
- **Deprioritize marital status**: Reduce emphasis on marital status as a segmentation factor.

--------------------------------------------------------------------------------


### **8. Outlier Detection**
**Observations:**
- Outliers in age and income suggest older, high-income customers purchase KP781.

**Recommendations:**
- Introduce **personalized offers** for high-value customers falling outside typical demographics (e.g., older but wealthy customers).
